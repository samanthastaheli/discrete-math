### Exercise 11.9


Represent the following graph as an adjacency list using a Python dictionary.

Then, create a function to determine whether a given edge is in the graph.

Next, create a function to return all the neighbors of a given vertex.

image.png

#### Your Answer:

```python
verticies = {'a','b','c','d','e','f'}
adjacency_list = {'a': ['b', 'c'],'b': ['a', 'c'],'c': ['a', 'b', 'd'],'d': ['c', 'e', 'f'],'e': ['d', 'f'],'f': ['d', 'e']}

def has_edge(edge, adjacency_list):
    if edge[0] in adjacency_list[edge[1]]:
      return True
    else:
      return False

def find_neighbors(vertex, adjacency_list):
    return adjacency_list[vertex]


edge1 = ['a', 'b'] # this edge is in the graph
edge2 = ['d', 'f'] # this edge is in the graph
edge3 = ['c', 'f'] # this edge is not in the graph

print(has_edge(edge1, adjacency_list)) # should be True
print(has_edge(edge2, adjacency_list)) # should be True
print(has_edge(edge3, adjacency_list)) # should be False

print(find_neighbors('a', adjacency_list)) # should print ['b', 'c']
print(find_neighbors('d', adjacency_list)) # should print ['c', 'e', 'f']
```
